```python
import pandas as pd

stats=dict(dp.comms.tolist())['stats']
logs = dict(dp.comms.tolist())['logs']
delays={}
averages={}
pathlengths={}
rx_packets=[]
for node,nlog in logs.items():
    delays[node]={}
    averages[node]={node:0}
    pathlengths[node]={node:0}
    for p in nlog['rx']:
        tx=p['source']
        if not delays[node].has_key(tx):
            delays[node][tx]=[]
            pathlengths[node][tx]=[]
        p['delay']=p['received']-p['time_stamp']
        delays[node][tx].append(p['delay'])
        pathlengths[node][tx].append(len(p['route']))
        rx_packets.append(p)
    for tx,lengths in pathlengths[node].items():
        averages[node][tx]=np.average(lengths)
    
def flatten(d):
    try:
        for v in d.itervalues():
            for nested_v in flatten(v):
                yield nested_v
    except AttributeError:
        for list_v in d:
            yield list_v
```

```python
fs=[lambda x : -x + 1,
    lambda x : -2*x +2 if x>0.5 else 2*x,
    lambda x : x]
sigmas=[0.0,0.5,1.0]
whitenized=lambda x : map(lambda f: f(x), fs)
white_class = lambda x: (whitenized(x).index(max(whitenized(x))))
#map(lambda f: plot(np.linspace(0,1,50),map(f,np.linspace(0,1,50)), label=fs), fs)
#legend()

observer = 'n0'
recommendation_nodes = ['n2','n3']
target = 'n1'
indirect_nodes = ['n4','n5']
t_direct = lambda x: 0.5 * max(whitenized(x)) * x
t_recommend = lambda x: 0.5 * (\
        2*len(recommendation_nodes) \
        /(2.0*len(recommendation_nodes)+len(indirect_nodes))) * max(whitenized(x)) * x
t_indirect = lambda x: 0.5 * (\
        2*len(indirect_nodes) \
        /(2.0*len(recommendation_nodes)+len(indirect_nodes))) * max(whitenized(x)) * x

def network_trust(t):
    t_sum=0
    who = [observer] + recommendation_nodes + indirect_nodes
    for whom in who:
        t_sum+=trust_inverted[whom][target][t]
    return t_sum/float(len(who))

def total_trust(t):
    Td = t_direct(trust_inverted[observer][target][t])
    Tr = average([t_recommend(trust_inverted[recommender][target][t]) for recommender in recommendation_nodes])
    Ti = average([t_indirect(trust_inverted[indirecter][target][t]) for indirecter in indirect_nodes])
    return sum((Td,Tr,Ti))

T_total = map(total_trust, range(1,dp.tmax/600))
T_network = map(network_trust, range(1,dp.tmax/600))
T_class = map(white_class, T_total)

# Trust Plots (Boxey)
from collections import OrderedDict
f, ax = plt.subplots(figsize=(13, 7))
ax.set_title("Trust Distributions for {} model against different T-models".format(dp.title))
ax.set_ylabel("Trust Value")
ax.set_xlabel("Trust Assessment")

_d={"t10": pd.Series(trust_inverted[observer][target]),
"t12": pd.Series(trust_inverted['n2'][target]),
"t13": pd.Series(trust_inverted['n3'][target]),
"t14": pd.Series(trust_inverted['n4'][target]),
"t15": pd.Series(trust_inverted['n5'][target]),
"t10-5": pd.Series(T_total),
"t10-net": pd.Series(T_network)}
plottables=OrderedDict(sorted(_d.items(), key=lambda k: k[0]))

with np.errstate(invalid='ignore'):
        sns.boxplot(
            pd.DataFrame.from_dict({ key: pd.Series(vals) for key,vals in plottables.items()}),
            ax=ax, showmeans=True, showbox=False, widths = 0.2, linewidth=2)
        ax.legend()
        ax.set_ylim(0,1.0)
save_chart(f,ax)
print np.e/np.pi
```
