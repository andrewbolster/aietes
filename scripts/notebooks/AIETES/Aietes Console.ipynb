```python
    def sphere(x, y, z, r=1.0):
        """
        Returns a sphere definition tuple (xs,ys,zs) for use with plot_wireframe
        """
        u, v = np.mgrid[0:2 * np.pi:10j, 0:np.pi:10j]
        xs = (r * np.cos(u) * np.sin(v)) + x
        ys = (r * np.sin(u) * np.sin(v)) + y
        zs = (r * np.cos(v)) + z

        return xs, ys, zs
```
